```python
import numpy as np
import pandas as pd
```

```python
estadisticas_BCRP = pd.read_excel("/content/estadisticas-100-anios-bcrp.xlsx")
estadisticas_BCRP.describe
```

```python
tcn_usa = estadisticas_BCRP.iloc[:, (21)]
fechas = estadisticas_BCRP.iloc[:, (14)]
```

```python
fechas
```

```python
tcn_usa
```

```python
tcn_usa_final = pd.concat([fechas, tcn_usa], axis=1)
```

```python
tcn_usa_final.columns = ["Fechas", "Tipo de Cambio Nominal USA"]
```

```python
tcn_usa_final
```

```python
tcn_usa_final = tcn_usa_final.apply(pd.to_numeric, errors='coerce')
threshold = 1e-1
tcn_usa_final[tcn_usa_final < threshold] = np.nan
tcn_usa_final = tcn_usa_final.dropna()
```

```python
tcn_usa_final
```

```python
tcn_usa_final['Fechas'] = tcn_usa_final['Fechas'].astype(int)
tcn_usa_final = tcn_usa_final.round(3)
```

```python
tcn_usa_final
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 6))
fig.set_dpi(100)
ax.plot(tcn_usa_final['Fechas'], tcn_usa_final['Tipo de Cambio Nominal USA'], color='red', linewidth=2.0)
ax.set_title('Tipo de Cambio Nominal USA a lo largo del tiempo', fontsize=20, pad=20, weight='bold')
ax.set_xlabel('Fechas', fontsize=10, labelpad=10, weight='bold')
ax.set_ylabel('Tipo de Cambio Nominal USA', fontsize=10, labelpad=10, weight='bold')
plt.show()
```

```python
def calcular_variacion_interanual(df):
    variaciones = []
    for i in range(1, len(df)):
        variacion = (df.iloc[i] - df.iloc[i-1]) / df.iloc[i-1] * 100
        variaciones.append(variacion)
    return variaciones
```

```python
def calcular_variacion_acumulada(df):
    variacion_acumulada = df.cumsum()
    return variacion_acumulada
```

```python

variaciones = calcular_variacion_interanual(tcn_usa_final['Tipo de Cambio Nominal USA'])
tcn_usa_final['Variación Interanual'] = [np.nan] + variaciones
tcn_usa_final['Variación Acumulada'] = calcular_variacion_acumulada(tcn_usa_final['Variación Interanual'])
tcn_usa_final
```

```python
df_variacion = tcn_usa_final[['Fechas', 'Variación Interanual']].copy()
desviacion_estandar = df_variacion['Variación Interanual'].std()
df_variacion = df_variacion.iloc[2:]
df_variacion
```

```python
df_mayor_desviacion = df_variacion[df_variacion['Variación Interanual'] > desviacion_estandar].copy()
plt.figure(figsize=(10, 6))
plt.plot(df_variacion['Fechas'], df_variacion['Variación Interanual'], color='red', linewidth=2.0)
plt.title('Variación Interanual del Tipo de Cambio Nominal USA', fontsize=20, pad=20, weight='bold')
plt.xlabel('Fechas', fontsize=15, weight='bold')
plt.ylabel('Variación Interanual (%)', fontsize=15, weight='bold')
plt.show()
```

```python
df2_variacion = df_variacion.iloc[5:]
variacion_promedio = df2_variacion['Variación Interanual'].mean()
variacion_promedio
```
